```python
# Libraries

import snscrape.modules.twitter as sntwitter
import pandas as pd
```

```python
# Scrape parameters

query = "COP26 since:2021-11-02_18:49:30_UTC until:2021-11-28 lang:en"
# Allows you to filter the twitter scrape for time, keywords, hashtags and a LOT more,
# see the image in 'technical notes' on teams for more information.

max_tweets = 100
# Once this many tweets are scraped, stop. Rule of thumb: 1hr = 100,000 tweets.

save_number = 10
# Save to .csv file after this many tweets are scraped, every time (in case you have to stop before the max.)

attributes_record = ["ID", "Date", "Likes", "Replies", "Retweets", "User ID", "Username", "Followers", "Verified", "Content"]
# Tweet data to save in the file. NOTE: also must be changed in --attributes_container.append([...])
```

```python
# Twitter Scrape Loop

# Created a list to append all tweet attributes(data)
attributes_container = []

for i,tweet in enumerate(sntwitter.TwitterSearchScraper("COP26 since:2021-11-02_18:49:30_UTC until:2021-11-28 lang:en").get_items()):
    if i>max_tweets:
        break
    if i%save_number==0:
        print(i)
        tweets_df = pd.DataFrame(attributes_container, columns=attributes_record)
        tweets_df.to_csv('Twitter Scrape 123.csv', index=False, columns=attributes_record)
    attributes_container.append([tweet.id, tweet.date, tweet.likeCount,
                                 tweet.replyCount, tweet.retweetCount, tweet.user.id,
                                 tweet.user.username, tweet.user.followersCount, tweet.user.verified*1,
                                 tweet.content])
    
# Creating a dataframe from the tweets list above
tweets_df = pd.DataFrame(attributes_container, columns=attributes_record)
tweets_df.sort_values(by='Date', ascending=True, inplace=True)
```

```python
# Read .csv file for analysis

tweets_df = pd.read_csv('Twitter Scrape 123.csv', usecols=attributes_record, date_parser = pd.to_datetime, parse_dates=["Date"])
```
